```python
import time
def 仿真移動(起始, 結束, duration=1.0):
    start_time = time.time()
    end_time = start_time + duration
    print(f"起始座標 ({起始[0]}, {起始[1]})")
    while time.time() < end_time:
        elapsed_time = time.time() - start_time
        progress = elapsed_time / duration
        current_x = int(起始[0] + (結束[0] - 起始[0]) * progress)
        current_y = int(起始[1] + (結束[1] - 起始[1]) * progress)
        print(f"移動座標 ({current_x}, {current_y})")
        time.sleep(0.01)  # 控制速度
    print(f"到達座標 ({結束[0]}, {結束[1]})")

仿真移動([100,100],[300,300],0.2)
```

```python
# 尋找最近的座標
import math

def 尋找最近的座標(x, y, points):
    min_distance = float('inf')
    closest_x = closest_y =  None
    for point in points:
        i, point_x, point_y = map(int, point.split(','))
        distance = math.sqrt((point_x - x) ** 2 + (point_y - y) ** 2)
        if distance < min_distance:
            min_distance = distance
            closest_x = point_x
            closest_y = point_y
    return closest_x, closest_y

# 中心座標
x = 497
y = 287
# 遍歷尋找 範圍座標 文字"Z" 顏色 相似度
posstr = job.dm.FindStrEx(359,260,635,315,"Z","f01010-000000|10f0f0-000000|f0f010-000000|10f010-000000|f010f0-000000|f0f0f0-000000|f08010-000000",1.0)
pos = posstr.split("|")
print(f'座標數組: {pos}')
closest_x, closest_y = 尋找最近的座標(x, y, pos) # 分割pos 並傳送尋找最近座標
print(f'最近的座標: {closest_x} - {closest_y}') # 輸出最近座標
```
