```python
def is_fullwidth(char):
    # 判斷字符是否為全形（通常 Unicode 編碼大於 U+FF00）
    if '\uFF00' <= char <= '\uFFEF':
        return True
    # 中文、日文、韓文等字符的 Unicode 範圍大於這個值，通常被認為是全形
    if '\u4e00' <= char <= '\u9fff':
        return True
    return False

def check_string(string,length=0):
    if length == 0:
        return string
    ss = 0
    for char in string:
        if is_fullwidth(char):
            ss += 1
        else:
            ss +=0.5
    while length > ss:
        if length - ss >= 1 :
            string += u'　'
            ss += 1
        else:
            string += ' '
            ss += 0.5
    return string

test_string = "信念意志頭盔(妖術師) (400218)"
print(check_string(f'{test_string}',30))
```

```python
import winreg
from datetime import datetime, timedelta

# 注册表路径
key_path = r'SOFTWARE\MyApp'
value_name = 'LastRunTime'

# 打开或创建注册表项
try:
    key = winreg.OpenKey(winreg.HKEY_CURRENT_USER, key_path, 0, winreg.KEY_ALL_ACCESS)
except FileNotFoundError:
    key = winreg.CreateKey(winreg.HKEY_CURRENT_USER, key_path)

# 获取上次运行时间
try:
    last_run_time_str, _ = winreg.QueryValueEx(key, value_name)
    last_run_time = datetime.strptime(last_run_time_str, '%Y-%m-%d %H:%M:%S')
    current_time = datetime.now()

    # 计算时间差
    time_diff = current_time - last_run_time
    if time_diff < timedelta(days=1):
        print("程序一天只能运行一次，请稍后再试！")
        winreg.CloseKey(key)
        exit()
except FileNotFoundError:
    print("这是程序第一次运行")

# 更新注册表中的时间
current_time_str = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
winreg.SetValueEx(key, value_name, 0, winreg.REG_SZ, current_time_str)
winreg.CloseKey(key)

# 程序主要逻辑
print("程序正在运行...")
```

```python
import ctypes
from ctypes import wintypes
import win32process
import win32gui
import win32api
import win32con
import psutil

psapi = ctypes.WinDLL('psapi')
kernel32 = ctypes.WinDLL('kernel32')

# 定义常量
PROCESS_QUERY_INFORMATION = 0x0400
PROCESS_VM_READ = 0x0010
MAX_PATH = 260

# 定义函数
GetWindowThreadProcessId = ctypes.windll.user32.GetWindowThreadProcessId
GetModuleFileNameExW = psapi.GetModuleFileNameExW
OpenProcess = kernel32.OpenProcess
CloseHandle = kernel32.CloseHandle

def get_exe_path_from_hwnd(hwnd):
    # 获取窗口对应的进程 ID
    pid = wintypes.DWORD()
    GetWindowThreadProcessId(hwnd, ctypes.byref(pid))

    # 打开该进程
    process_handle = OpenProcess(PROCESS_QUERY_INFORMATION | PROCESS_VM_READ, False, pid.value)
    if not process_handle:
        return None

    # 获取模块的文件名（即可执行文件路径）
    exe_path_buffer = ctypes.create_unicode_buffer(MAX_PATH)
    if GetModuleFileNameExW(process_handle, None, exe_path_buffer, MAX_PATH) > 0:
        exe_path = exe_path_buffer.value
    else:
        exe_path = None

    # 关闭句柄
    CloseHandle(process_handle)

    return exe_path

# 示例用法：获取当前活动窗口的路径
hwnd = win32gui.GetForegroundWindow()  # 或替换为你感兴趣的窗口句柄
exe_path = get_exe_path_from_hwnd(988554)

if exe_path:
    print(f"Executable path: {exe_path}")
else:
    print("Unable to retrieve the executable path.")
```
